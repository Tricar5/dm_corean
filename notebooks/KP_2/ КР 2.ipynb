```python
# 5 Преобразование переменных
import numpy as np

def read_list_from_string(text: str):
    """Функция для преобразования строчки с закодированым массивом внутри обратно в список"""
    text = text[1:-1]
    arr = text.split(',')
    return [el.strip() for el in arr]

# Стратегия

data['Listed_topicDetails.topicCategories'] = data['topicDetails.topicCategories'].apply(lambda x: read_list_from_string(x) if isinstance(x, str) else np.nan)

data['Число_категорий_в_topicDetails.topicCategories'] = data['Listed_topicDetails.topicCategories'].apply(lambda x: len(x) if isinstance(x, list) else np.nan)


data['Число_слов_в_brandingSettings.channel.keywords'] = data['brandingSettings.channel.keywords'].apply(lambda x: len(x.split()) if isinstance(x, str) else np.nan)

display(data[['topicDetails.topicCategories', 'Listed_topicDetails.topicCategories', 'Число_категорий_в_topicDetails.topicCategories', 'brandingSettings.channel.keywords', 'Число_слов_в_brandingSettings.channel.keywords']].sample(10))
```

## Задача 5. Постарайтесь письменно сформулировать для них содержательную гипотезу в терминах КРИВОлинейной связи: её наличия|отсутствия в целом и между отдельными категориями. Также письменно сформулируйте релевантную статистическую гипотезу.


```python
# 14 Проверка на аппроксимацию с помощью кривой
fig = px.scatter(data, x=vars[0], y=vars[1], trendline='lowess')

fig.show()
```

Мы пришли к тому, что  между признаками отсутствуте линейная связь. А также графически криволинейная связь не подтверждается (аппроксимированная кривая по Lowess не выделяет пиков средних значений)

Соответственно, мы можем проверить наличие связи между группами с помощью кросс-таблиц и $\chi ^{2} $

**Содержательная гипотеза:**

Каналы имеющие разное количество заявленных тематик имеют разное

**Статистическая гипотеза:**


H0: $\chi ^{2}  = 0 $ в ГС
H1: $\chi^{2}  != 0 $ в ГС

```python
# 14 формируем группы для создания таблицы и проверки связи
inv_borders = [
                       np.nanpercentile(data['Число_слов_в_brandingSettings.channel.keywords'], 25),
                       np.nanpercentile(data['Число_слов_в_brandingSettings.channel.keywords'], 50),
                       np.nanpercentile(data['Число_слов_в_brandingSettings.channel.keywords'], 75),
                       np.nanpercentile(data['Число_слов_в_brandingSettings.channel.keywords'], 100)]


def make_group_from_borders(word_count):
    if word_count < inv_borders[0]:
        return '1 Low word_count'
    elif inv_borders[0] <= word_count < inv_borders[1]:
        return '2 Lower medium word_count'
    elif inv_borders[1] <= word_count < inv_borders[2]:
        return '3 Higher medium word_count'
    else:
        return '4 High word_count'

data['word_count.group'] = data['Число_слов_в_brandingSettings.channel.keywords'].apply(make_group_from_borders)
```

```python
# 16 выводим статистику по группам

from randan.descriptive_statistics import NominalStatistics

var = 'word_count.group'

ns = NominalStatistics(data, [var])

freq_table = ns.frequencies()[var]
```

## Задача 6. Проверьте эти гипотезы математически. Строго соблюдайте ограничения применяемых методов. Проинтерпретируйте результат.

```python
# 16 кросстаблица
from randan.bivariate_association import Crosstab

ctab = Crosstab(data, 'Число_категорий_в_topicDetails.topicCategories', 'word_count.group')
```

```python
# 17 Оценка значимости
print('P-value', ctab.pvalue)
print('Level of Significance', ctab.sig_level)

ctab.pvalue < ctab.sig_level
```

### Интерпретация

Основываясь на данных теста и наблюдаемых частотах, мы не имеем оснований отклонить нулевую гипотезу (p-value > level of significance)
