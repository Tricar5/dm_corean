# Relevance channel estimation

```python
# 1 импорт базовых библиотек и конфига
import pathlib

import pandas as pd
import yaml
ROOT = pathlib.Path().resolve().parent.parent

with open(ROOT / "config.yml", "r") as f:
    config = yaml.safe_load(f)
```

```python
# 2 установка переменных

DATA_PATH = ROOT / config['api']['DATA_PATH']

#query = "корейская косметика"
#type_query = "video"

# Для поиска видео с учетом даты
publishedAfter = '2020-01-01T00:00:00Z'
publishedBefore = '2022-01-01T00:00:00Z'
```

```python
# 3  загрузка датафрейма

df = pd.read_excel(DATA_PATH / 'seminar4' / 'Channel_characteristics.xlsx')
display(df)
```

```python

len(df["id"].unique())
```

```python
# 6 Селекция по количеству подписчиков
import numpy as np


print('Mean', np.mean(df['statistics.subscriberCount']))

print('Min', np.min(df['statistics.subscriberCount']))
print('25', np.nanpercentile(df['statistics.subscriberCount'].values, q=25))
print('50', np.nanpercentile(df['statistics.subscriberCount'], q=50))
print('75', np.nanpercentile(df['statistics.subscriberCount'], q=75))
print('90', np.nanpercentile(df['statistics.subscriberCount'], q=90))
print('95', np.nanpercentile(df['statistics.subscriberCount'], q=95))
print('Max', np.max(df['statistics.subscriberCount']))

# Отбираем по 75% количества подписчиков

choosen_lvl = np.nanpercentile(df['statistics.subscriberCount'], q=75)
```

```python
# 4 формирование списка каналов
df.columns
```

```python
# 5 Отбор индексирируемых колонок в датафрейме для отбора

col_list = ['id', 'snippet.thumbnails.default.url','snippet.title','snippet.description','topicDetails.topicCategories', 'brandingSettings.channel.keywords']

selected_df = df[df['statistics.subscriberCount'] >= choosen_lvl][col_list].copy()
selected_df
```

```python
# 6 выделение тестовых индексов

valid_indexes = selected_df.sample(50, random_state=42).index
valid_indexes
```

```python
# 7 отбор для разметки

selected_df['Вероятностный отбор'] = 0
selected_df.loc[valid_indexes, 'Вероятностный отбор'] = 1
```

```python
selected_df['Вероятностный отбор'].value_counts()
```

```python
import os

if not os.path.exists(DATA_PATH / "seminar5"):
    os.makedirs(DATA_PATH / "seminar5")


selected_df.to_excel(DATA_PATH / 'seminar5' / 'Каналы_вероятностный_отбор.xlsx', index=False)
```
