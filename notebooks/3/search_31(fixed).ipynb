```python
import pathlib
import yaml
ROOT = pathlib.Path().resolve().parent.parent

with open(ROOT / "config.yml", "r") as f:
    config = yaml.safe_load(f)
```

```python
# 3
DATA_PATH = config['api']['DATA_PATH']

query = "корейская косметика"
type_query = "channel"
```

```python
API_KEY_STORAGE = config['api']['keys']

key_iterator = iter(API_KEY_STORAGE)

API_KEY = next(key_iterator)

print(API_KEY)
```

```python
# 4
import googleapiclient.discovery as api
```

```python
# 5
api_service_name = "youtube"
api_version = "v3"
    
youtube = api.build(api_service_name,
                        api_version, 
                       developerKey = API_KEY)
```

```python
# 6
request = youtube.search().list(
        part="snippet",
        maxResults=50,
        q=query,
        regionCode="RU",
        type=type_query
)
response = request.execute()
```

```python
# 7
display(response, '---', response.keys(), '---', response['pageInfo']['totalResults'])
```

```python
#8
import pandas as pd
```

```python
# 9
df = pd.json_normalize(response['items'])
df
```

```python
# 10 
request = youtube.search().list(
        part ="snippet",
        maxResults=50,
        q = query,
        regionCode ="RU",
        type = type_query,
     pageToken = response['nextPageToken']
)
response = request.execute()
```

```python
# 11
df_additional = pd.json_normalize(response['items'])
df_additional
```

```python
# 12 альтернатива
# Оформление выдачи чанка #6 в таблицу 

df_supplemented = df.copy(deep=True)

del df

df_supplemented
```

```python
#13
i = 0
while 'nextPageToken' in response.keys():
    request = youtube.search().list(
        part ="snippet",
        maxResults=50,
        q = query,
        regionCode ="RU",
        type = type_query,
        pageToken = response['nextPageToken']
    )
    response = request.execute()  

    print(f'Итерация №{i}')
    i += 1
    
    df_additional = pd.json_normalize(response['items'])
    df_supplemented = pd.concat([df_supplemented, df_additional])
df_supplemented
```

```python
#14
# Удаление дубликатов каналов и запись в Excel:
df_supplemented.index = range(1,len(df_supplemented)+1)
df_supplemented = df_supplemented.drop_duplicates('id.channelId')
display(df_supplemented)
```

```python
# 15

import os


if not os.path.exists(ROOT / DATA_PATH / "seminar3"):
    os.makedirs(ROOT / DATA_PATH / "seminar3")

df_supplemented.to_excel(ROOT / DATA_PATH / 'Каналы_No_sorting.xlsx')
```
